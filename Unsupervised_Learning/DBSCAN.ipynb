#### Birleştirici (Agglomerative) - Hiyerarşikal(Hierarhical) - DBSCAN

```python
import mglearn
mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
X,y = make_blobs(random_state=42)
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
k = agg.fit_predict(X)
mglearn.discrete_scatter(X[:,0],X[:,1],k)
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
X,y = make_blobs(random_state=0,n_samples=12)
linkange_array = ward(X)
```

```python
dendrogram(linkange_array)
```

### DBSCAN
* verideki yoğun bölgeleri bularak kümeleme yapar. Seyrek bölgeleri parazit olarak algılar.
- Yoğun bölgeden rastgele bir nokta seçilir ve merkez yapılır.
- Ardından belirlenen yarıçapın (Eps parametresi) kapsadığı alandaki noktalar kümelenir.

```python
from sklearn.cluster import DBSCAN
X,y = make_blobs(random_state=0,n_samples=12)
dbscan = DBSCAN()
```

```python
cluster = dbscan.fit_predict(X)
cluster
```

* Yukarıdaki -1 etiketi paraziti ifade eder.

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=200, noise=0.06, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN()
cluster = dbscan.fit_predict(X_scaled)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=cluster,cmap=mglearn.cm2,s=60)
```
