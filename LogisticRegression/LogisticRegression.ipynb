```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
x_train,x_test,y_train,y_test = train_test_split(data.data,data.target,stratify=data.target,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='liblinear').fit(x_train,y_train)
```

```python
print(f'Train Accucary {logreg.score(x_train,y_train)}')
print(f'Test Accuracy {logreg.score(x_test,y_test)}')
```

```python
from sklearn.linear_model import LogisticRegression
logreg100 = LogisticRegression(C=100,solver='liblinear').fit(x_train,y_train)

print(f'Train Accucary {logreg100.score(x_train,y_train)}')
print(f'Test Accuracy {logreg100.score(x_test,y_test)}')
```

```python
from sklearn.linear_model import LogisticRegression
logreg001 = LogisticRegression(C=0.01,solver='liblinear').fit(x_train,y_train)

print(f'Train Accucary {logreg001.score(x_train,y_train)}')
print(f'Test Accuracy {logreg001.score(x_test,y_test)}')
```

```python
for C, market in zip([0.001,1,100],['o','^','v']):
    lr_l1=LogisticRegression(penalty='l1',max_iter=1000,solver='liblinear',C=C).fit(x_train,y_train)
    print(f'For C={C}, train accuracy= {lr_l1.score(x_train,y_train)}')
    print(f'For C={C}, test accuracy= {lr_l1.score(x_test,y_test)}')
```

```python
from sklearn.datasets import make_blobs
import mglearn
```

```python
%matplotlib inline
```

```python
X,y = make_blobs(random_state=42)
import matplotlib.pyplot as plt
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.legend()
```

Linear Support Vector (LinearSVC)

```python
from sklearn.svm import LinearSVC
linear_svm = LinearSVC().fit(X,y)
linear_svm.coef_
```

```python
import numpy as np
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)
for coef, intercept, color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
    plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
plt.ylim(-10,15)
plt.xlim(-10,8)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.legend(['class1','class2','class2','correct class 0','correct class1','correct class2'],loc=(1.02,0.4))
```
