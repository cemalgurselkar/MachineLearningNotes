## Manifold Learning

* Linear olmayan veri setleri için kullanılır. Amaç boyut indirgeme

```python
# t-SNE Tekniği: 
# Veriler arası mesafeyi en iyi şekilde koruyarak boyut indirgeme yapmak.

from sklearn.datasets import load_digits
data = load_digits()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(data.data)
```

```python
digits_pca = pca.transform(data.data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
color = ['#476A2A','#7851B8','#BD3430','#4A2D4E','#875525',
        '#A83683','#4E655E','#853541','#3A3120','#535D8E']

plt.figure(figsize=(10,10))
plt.xlim(digits_pca[:,0].min(),digits_pca[:,0].max())
plt.ylim(digits_pca[:,0].min(),digits_pca[:,0].max())

for i in range(len(data.data)):
    plt.text(digits_pca[i,0],digits_pca[i,1],str(data.target[i]),
             color=color[data.target[i]],
             fontdict={'weight':'bold','size':9})
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
```

```python
digits_tsne = tsne.fit_transform(data.data)
```

```python
color = ['#476A2A','#7851B8','#BD3430','#4A2D4E','#875525',
        '#A83683','#4E655E','#853541','#3A3120','#535D8E']

plt.figure(figsize=(10,10))
plt.xlim(digits_tsne[:,0].min(),digits_tsne[:,0].max())
plt.ylim(digits_tsne[:,0].min(),digits_tsne[:,0].max())

for i in range(len(data.data)):
    plt.text(digits_tsne[i,0],digits_tsne[i,1],str(data.target[i]),
             color=color[data.target[i]],
             fontdict={'weight':'bold','size':9})
```

* Yukarıda görüldüğü gibi sınıflar güzel bir şekilde ayrıldı.
* Manifold teknikleri: düzensiz verileri aralarındaki ilikşiyi koruyarak boyut indirgeme yapmak.
