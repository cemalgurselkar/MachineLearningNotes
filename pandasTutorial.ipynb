## Pandas Row-Column Transformations

```python
import seaborn as sns
import pandas as pd
import numpy as np
tips = sns.load_dataset('tips')
tips.head(6)
```

```python
tips_bg = tips.groupby(['day','sex']).agg(({'size':'sum'}))
tips_bg
```

```python
tips_us = tips_bg.unstack()
tips_us
```

```python
tips_us.columns
```

```python
tips_us.stack()
```

```python
tips_us.stack(0)
```

```python
tips_us.stack(1)
```
